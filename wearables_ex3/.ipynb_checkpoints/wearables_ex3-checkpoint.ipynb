# 1. Segmenting

We know that the signals' sampling frequency is 200Hz, so utilizing this information, we can divide the data into 10s segments, because 200 samples/s = 2000 samples/10s

```python
def segmentor(data, fr, segment_length):
    x = 0
    segments = []
    while x<len(data):
        segment = data.loc[x:(x+fr*segment_length-1)].reset_index(drop=True)
        segments.append(segment)
        x = x + len(segment)
    return segments
```

```python
segments_ecg = segmentor(df_ecg['ecg_sig'], 200, 10)
segments_ppg = segmentor(df_ppg['ppg_sig'], 200, 10)

print(segments_ecg[0])
```

```python
def average_waveform(segments, fr, time):
    peak_is = []
    peaks = []
    z = 0
    average_peak_locations = []
    for i in segments:
        peak_is.append(sp.signal.find_peaks(i, height=1.00))
        peaks_at_z =[]
        peak_locs = []
        for j in peak_is[z][0]:
            peak_locs.append(j)
            peaks_at_z.append(segments[z][j])
        peaks.append(peaks_at_z)
        average_peak_locations.append(peak_locs)
        z+=1
    peaks.remove(peaks[len(peaks)-1])
    
    df1 = pd.DataFrame(peaks)
    df2 = pd.DataFrame(average_peak_locations)
    indices_mean = df2.mean(axis=0).astype(int)
    peaks_mean = df1.mean(axis=0)
    
    average_waveform = []
    x = 0
    for i in range(0, fr*time):
        if x<len(indices_mean):
            if i == indices_mean.loc[x]:
                average_waveform.append(peaks_mean[x])
                x+=1
            else:
                average_waveform.append(0)
        else:
            average_waveform.append(0)
        
    return average_waveform   
```

```python
def average_waveform_calculator(segments, fr, time, r_height):
    peak_is = []
    peaks = []
    hr = []
    intervals = []
    beats = []
    timeline = np.array(list(range(0,fr*time)))
    for i in segments:
        r_peaks = sp.signal.find_peaks(i, height=r_height)
        #print(r_peaks)
        peak_is.append(r_peaks[0])
        peaks.append(r_peaks[1])
        RR_intervals = np.diff(timeline[r_peaks[0]])
        heart_rate = 60.0/(np.mean(RR_intervals/fr))
        hr.append(heart_rate)
        intervals.append(np.mean(RR_intervals))
        
    mean_hr = math.floor(np.mean(hr))
    mean_intervals = math.floor(np.mean(intervals)/2)
    print(mean_hr)
    print(mean_intervals)
    
    z = 0
    for i in peak_is:
        a = 0
        segment_beats = []
        for j in i:
            x = segments[z][peak_is[z][a]-mean_intervals:peak_is[z][a]]
            y = segments[z][peak_is[z][a]:peak_is[z][a]+mean_intervals]
            segment_beats = pd.concat([x, y], axis=0).reset_index(drop=True)
            beats.append(segment_beats)
            #print(peak_is[z][a])
            a+=1
        #print('STOP')
        z+=1
        
    #print(beats[1])
    
    df1 = pd.DataFrame(beats).transpose()
    df1_mean = df1.mean(axis=1)
    #print(df1_mean.head())
    
    timeframe = list(range(0, len(df1_mean)))
    #plt.figure(figsize=(10,7))
    #plt.plot(timeframe, df1_mean)
    #plt.show()
    
    #new_segments = segmentor(df1_mean, fr, 1)
    #df2 = pd.DataFrame(new_segments).transpose()
    #df2_mean = df2.mean(axis=0)
    
    return timeframe, df1_mean
```

```python
#First ECG
ecg_avg = average_waveform_calculator(segments_ecg, 200, 10, 12000)

#For PPG, I left out the last segment, as it messed the whole thing up and this is a very much hard-coded 
#fix for the issue, but it works so...
ppg_avg = average_waveform_calculator(segments_ppg[0:len(segments_ppg)-1], 200, 10, [20, 100])
```
